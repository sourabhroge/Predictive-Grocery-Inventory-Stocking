Everything looks fine for the three datasets apart from the timestamp column in each dataset.
Using the helper function we'll convert this to the correct type for each dataset

```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):

  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%d-%m-%Y %H:%M')
  return dummy
```

```python
stock_df = convert_to_datetime(stock_df, 'timestamp')
stock_df.info()
```

MERGE DATA
Currently we have 3 datasets. In order to  include all of this data within a predictive model, 
we need to merge them together into 1 dataframe.

The client indicates that they want the model to predict on an hourly basis. Looking at the data model, 
we can see that only column that we can use to merge the 3 datasets together is timestamp.
So, we must first transform the timestamp column in all 3 datasets to be based on the hour of the day,
then we can merge the datasets together.

```python
sales_df.head()
```

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%d-%m-%Y %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%d-%m-%Y %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
```

```python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
```

```python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
```

As we wanted the predective modelling on hourly basis, the minutes and seconds in the timestamp columns have been reduced to 00.
The next thing to do, is to aggregate the datasets in order to combine rows which have the same value for timestamp.

For the sales data, we want to group the data by timestamp but also by product_id. 
When we aggregate, we must choose which columns to aggregate by the grouping. For now, let's aggregate quantity.

```python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
```

For the stock data, we want to group it in the same way and aggregate the estimated_stock_pct.

```python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
```

Finally, for the temperature data, product_id does not exist in this table, so we simply need to group by timestamp and
aggregate the temperature.

```python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

This gives us the average temperature of the storage facility where the produce is stored in the warehouse by unique hours 
during the week. Now, we are ready to merge our data. 
We will use the stock_agg table as our base table, and we will merge our other 2 tables onto this.

```python
merged_df = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged_df
```

```python
merged_df = merged_df.merge(temp_agg, on='timestamp', how='left')
merged_df.head()
```

```python
merged_df.info()
```

As we can see from the information that we get from merged dataframe we have some null values in the quantity column of the 
dataset. For performing further operations these null values need to be treated first.
As the null values are from quantity column we can say that 0 sales were made for that product within this hour.
So we can replace these null values with 0 after consulting the same with the client side. 

```python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
```

```python
product_categories = sales_df[['product_id', 'category']]
product_categories = product_categories.drop_duplicates()

product_price = sales_df[['product_id', 'unit_price']]
product_price = product_price.drop_duplicates()
```

```python
merged_df = merged_df.merge(product_categories, on="product_id", how="left")
merged_df.head()
```

```python
merged_df = merged_df.merge(product_price, on="product_id", how="left")
merged_df.head()
```

```python
merged_df.info()
```

FEATURE ENGINEERING
We have our cleaned and merged data. Now we must transform this data so that the columns are in a suitable format for a machine
learning model. In other terms, every column must be numeric. There are some models that will accept categorical features, 
but for this exercise we will use a model that requires numeric features.

At first let's start with timestamp. In it's current form it's not very useful so we will split it into day of month, day of 
week and hour.


```python
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
```

The next column we will engineer is category. Currently it is in categorical form and we will convert it into numerical form 
by creating dummy variables from this categorical column.


```python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

Looking at the above info, we can see that product_id is the only column that is not numeric.

Since each row represents a unique combination of product_id and timestamp by hour, and the product_id is simply an ID column,
it will add no value by including it in the predictive model. Hence, we shall remove it from the modeling process.

```python
merged_df.drop(columns=['product_id'], inplace=True)
merged_df.head()
```

MODELLING

```python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

```python
K = 10
split = 0.75
```

Now let's create a loop to train K models with a 75/25% random split of the data each time between training and test samples.

```python
accuracy = []

for fold in range(0, K):

  # Instantiate algorithm
  model = RandomForestRegressor()
  scaler = StandardScaler()

  # Create training and test samples
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)

  # Scale X data, we scale the data because it helps the algorithm to converge
  # and helps the algorithm to not be greedy with large values
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)

  # Train model
  trained_model = model.fit(X_train, y_train)

  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)

  # Compute accuracy, using mean absolute error
  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
```

```python
features = [i.split("__")[0] for i in X.columns]
importances = model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

From the above table it is evident that unit price and temperature were more important in predicting stock.
It is also seen that the time of the day was also important in predicting stock.
Product wise categorisation wasn't much helpful.

With these insights, we can report back to the client regarding the live time predicting of stock with respect to sales.
